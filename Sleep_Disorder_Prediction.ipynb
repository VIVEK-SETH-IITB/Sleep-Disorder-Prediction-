```python
#importing the libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#loading the dataset
df = pd.read_csv('Sleep_health_and_lifestyle_dataset.csv')
df.head()
```

# Preprocessing

```python
#checking for missing values
df.isnull().sum()
```

```python
#replacing the null values with 'None' in the column 'Sleep Disorder'
df['Sleep Disorder'].fillna('None', inplace=True)
```

```python
#drop column Person ID
df.drop('Person ID', axis=1, inplace=True)
```

```python
#checking the number of unique values in each column
print("Unique values in each column are:")
for col in df.columns:
    print(col,df[col].nunique())
```

#Splitting the blood pressure into systolic and diastolic

```python
#spliting the blood pressure into two columns
df['systolic_bp'] = df['Blood Pressure'].apply(lambda x: x.split('/')[0])
df['diastolic_bp'] = df['Blood Pressure'].apply(lambda x: x.split('/')[1])
#droping the blood pressure column
df.drop('Blood Pressure', axis=1, inplace=True)
```

```python
#replacing normal weight with normal in BMI column
df['BMI Category'] = df['BMI Category'].replace('Normal Weight', 'Normal')
```

```python
df.head()
```

#Checking the unique values from each categorical column

```python
#unique values from categorical columns
print(df.Occupation.unique())
print('\n')
print(df['BMI Category'].unique())
print('\n')
print(df['Sleep Disorder'].unique())
```

#Explorative Data Analysis

```python
fig,ax = plt.subplots(3,3,figsize=(20,10))
sns.countplot(x = 'Gender', data = df, ax = ax[0,0])
sns.histplot(x = 'Age', data = df, ax = ax[0,1], bins = 10)
sns.histplot(x = 'Sleep Duration', data = df, ax = ax[0,2], bins = 10)
sns.countplot(x = 'Quality of Sleep', data = df, ax = ax[1,0])
sns.histplot(x = 'Physical Activity Level', data = df, ax = ax[1,1], bins = 10)
sns.countplot(x = 'Stress Level', data = df, ax = ax[1,2])
sns.countplot(x = 'BMI Category', data = df, ax = ax[2,0])
sns.histplot(x = 'Daily Steps', data = df, ax = ax[2,1], bins = 10)
sns.countplot(x = 'Sleep Disorder', data = df, ax = ax[2,2])
```

$Observation:$


The gender distribution is nearly balanced, with a roughly equal number of males and females. The majority of individuals fall within the 30-45 age range. A significant portion of the population rates their sleep quality above 5, indicating satisfactory sleep duration. Furthermore, a substantial proportion of people exhibit a normal BMI, which is closely associated with the prevalence of sleep disorders. This suggests an approximately equal count of individuals with and without sleep disorders.

```python
#Gender count plot
sns.countplot(x = 'Gender', data = df, palette = 'hls', hue = 'Sleep Disorder').set_title('Gender and Sleep Disorder')
```

$Observation:$


The majority of both males and females do not experience any sleep disorders. Nevertheless, a higher incidence of sleep disorders is observed among females compared to males. Specifically, there is a notable prevalence of Sleep Apnea among females in contrast to males. Conversely, a greater number of males are affected by Insomnia compared to females.

#Effect of Occupation on Sleep Disorder

```python
ax = sns.countplot(x = 'Occupation', data = df, hue = 'Sleep Disorder')
ax.set_xticklabels(ax.get_xticklabels(), rotation = 90)
```

$Observation:$


The graph provides a clear illustration of how occupation significantly influences the occurrence of sleep disorders. Among various occupations, nurses exhibit a higher susceptibility to Sleep Apnea, with only a minimal portion being devoid of sleep disorders. Following nurses, the Salesperson occupation presents the highest incidence of Insomnia, with teachers ranking next. However, some professions stand out for their low occurrence of both Sleep Apnea and Insomnia, including Engineers, Doctors, Accountants, and Lawyers.

While the numbers of Software Engineers and Managers are limited, limiting detailed conclusions, the Sales Representative occupation demonstrates a sole instance of Sleep Apnea without any cases of Insomnia or other sleep disorders.

#BMI and Sleep Disorder

#Label Encoding for categorical variables

```python
from sklearn import preprocessing
label_encoder = preprocessing.LabelEncoder()
```

```python
vars = ['Gender', 'Occupation','BMI Category','Sleep Disorder']
for i in vars:
    label_encoder.fit(df[i].unique())
    df[i] = label_encoder.transform(df[i])
    print(i,':' ,df[i].unique())
```

#Correlation Matrix Heatmap

```python
#Correlation Matrix Heatmap
plt.figure(figsize=(20, 16))
sns.heatmap(df.corr(), annot = True, cmap = 'coolwarm')
```

```python
#Train Test Split

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(df.drop('Sleep Disorder',axis=1), df['Sleep Disorder'], test_size=0.3, random_state=42)
```

# Model Building

$Decision Tree Classifier$

```python
from sklearn.tree import DecisionTreeClassifier
dtree = DecisionTreeClassifier()
dtree
```

```python
#Training the model with train dataset
dtree.fit(X_train, y_train)
```

```python
#Training accuracy
print("Training Accuracy:",dtree.score(X_train,y_train))
```

# Evaluation

```python
d_pred = dtree.predict(X_test)
d_pred
```

# Using Confusion matrix heatmap to visualize the model accuracy

```python
from sklearn.metrics import confusion_matrix
sns.heatmap(confusion_matrix(y_test, d_pred), annot=True, cmap='Blues', fmt='g')
plt.title('Confusion Matrix')
plt.xlabel('Actual')
plt.ylabel('Predicted')
plt.show()
```

$Observation:$


The boxes along the diagonal represent the number of accurate positive outcomes, indicating correct predictions generated by the model. Conversely, the boxes off the diagonal display the quantity of inaccurate positive outcomes, denoting incorrect predictions made by the model.

# Dsitribution plot for predicted and actual values

```python
ax = sns.distplot(y_test, hist=False, color="r", label="Actual Value")
sns.distplot(d_pred, hist=False, color="b", label="Fitted Values" , ax=ax)
plt.title('Actual vs Fitted Values for Sleep Disorder Prediction')
plt.xlabel('Sleep Disorder')
plt.ylabel('Proportion of People')
plt.show()
```

In the visualization, red indicates the real values while blue corresponds to the predicted values. The graph illustrates that the model's predictions somewhat track the curve of the real values; however, there remains a discrepancy between the predicted and actual values. This discrepancy indicates that the model's predictions are not yet accurate.

# Classification Report

```python
from sklearn.metrics import classification_report
print(classification_report(y_test, d_pred))
```

$Observation:$

The model performs admirably, boasting an accuracy of 87% and an average F1 score of 0.83. It effectively anticipates the presence of sleep disorders with a high degree of precision.

# Random Forest Classifier

```python
from sklearn.ensemble import RandomForestClassifier
rfc = RandomForestClassifier(n_estimators=100, random_state=42)
```

```python
#Training the model with train dataset
rfc.fit(X_train, y_train)
```

```python
#Training accuracy
print("Training accuracy: ",rfc.score(X_train,y_train))
```

#Evaluation

```python
rfc_pred = rfc.predict(X_test)
rfc_pred
```

```python
#Using confusion matrix heatmap to visualize the model accuracy

#confusion matrix heatmap
sns.heatmap(confusion_matrix(y_test, rfc_pred), annot=True, cmap='Blues')
plt.title('Confusion Matrix')
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.show()
```

$Observation:$

The accuracy of the Random Forest Classifier model surpasses that of the Decision Tree Classifier model. The diagonal boxes represent True Positives, indicating accurate predictions, while the off-diagonal boxes indicate the count of false positive results, representing incorrect predictions by the model. The model's proficiency in making accurate predictions is evident from the lower count of false positive values, indicating its effectiveness in delivering correct results.

```python
# Distribution plot for predicted and acutal values
ax = sns.distplot(y_test, hist=False, color="r", label="Actual Value")
sns.distplot(rfc_pred, hist=False, color="b", label="Predicted Values" , ax=ax)
plt.title('Actual vs Predicted values for Sleep Disorder')
plt.xlabel('Sleep Disorder')
plt.ylabel('Proportion of Patients')
plt.show()
```

$Observation:$

The Random forest classifier has improved accuracy as compared to the Decision Tree which is shown with the gap between the actual and predcited values which was wider incase of Descision Tree Classifier.

# Classification Report

```python
print(classification_report(y_test, rfc_pred))
```

$Observation:$

The Random Forest Classifier model has an accuracy of 89%  and an avergae F1 score of 0.86. From the metrics it is quite clear that the model is able to predict the sleep disorder quite effectively, with increased accuracy than Decision Tree Classifer.
